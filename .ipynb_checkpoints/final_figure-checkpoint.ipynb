```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
%matplotlib qt

size = 9
plt.rcParams.update({
    # --- Global font ---
    'font.family': 'Arial',            # or 'Helvetica', 'DejaVu Sans', 'Times New Roman'
    'font.size': size,                    # default text size for everything

    # --- Axis labels & title ---
    'axes.labelsize': size,
    'axes.titlesize': size,

    # --- Tick labels ---
    'xtick.labelsize': size,
    'ytick.labelsize': size,
    'xtick.direction': 'in',
    'ytick.direction': 'in',

    # --- Legend ---
    'legend.fontsize': size,
    'pdf.fonttype': 42,
    'ps.fonttype': 42
    
})

# ── 1️⃣ Figure and outer GridSpec ─────────────────────────────
fig = plt.figure(figsize=(5.2,6))
outer = gridspec.GridSpec(
    3, 1,
    hspace=0.2,    # vertical gap between major rows
    top=0.95, bottom=0.07, left=0.05, right=0.95
)

axes = []

# ── 2️⃣ Top row: 2 subplots ───────────────────────────────────
top_grid = gridspec.GridSpecFromSubplotSpec(1, 2, subplot_spec=outer[0], width_ratios=[1,1.5], wspace=0.05)

# Left subplot (single)
ax_left = fig.add_subplot(top_grid[0, 0])
ax_left.spines['top'].set_visible(False)
ax_left.spines['right'].set_visible(False)
ax_left.spines['bottom'].set_visible(False)
ax_left.spines['left'].set_visible(False)
ax_left.set_xticks([])
ax_left.set_yticks([])
axes.append(ax_left)

# Right subplot (nested 2×3 grid)
nested_top = gridspec.GridSpecFromSubplotSpec(
    2, 3, subplot_spec=top_grid[0, 1], hspace=0.4,
    wspace=0.5
)

# common x, y axis labels for the nested 2x3 block
big_ax = fig.add_subplot(top_grid[0, 1], frameon=False)
big_ax.tick_params(labelcolor='none', top=False, bottom=False, left=False, right=False)
big_ax.set_xticks([])
big_ax.set_yticks([])
big_ax.set_xlabel("Time (s)", fontsize=size, labelpad=18)
big_ax.set_ylabel("Rate (Hz)", fontsize=size, labelpad=15)

# ── 3️⃣ UPDATED helper function — 3 sections now ───────────────
def make_nonuniform_row(subplot_spec, width):
    """
    Creates a row with:
       - left:   2 tightly spaced subplots
       - mid:    1 subplot
       - right:  1 NEW subplot
    Total = 4 subplots per row (2 + 1 + 1)
    """
    outer_row = gridspec.GridSpecFromSubplotSpec(
        1, 3, subplot_spec=subplot_spec,
        width_ratios=[1.5, 1, 1],   # give balanced room for new panel
        wspace=width                  # spacing between main blocks
    )

    # Left section → 2 tightly spaced subplots
    left = gridspec.GridSpecFromSubplotSpec(
        1, 2, subplot_spec=outer_row[0], wspace=0.4
    )

    # Middle subplot (existing)
    mid = gridspec.GridSpecFromSubplotSpec(
        1, 1, subplot_spec=outer_row[1]
    )

    # NEW rightmost subplot
    right_new = gridspec.GridSpecFromSubplotSpec(
        1, 1, subplot_spec=outer_row[2]
    )

    return left, mid, right_new

# ── 4️⃣ Middle row: 4 subplots now ─────────────────────────────
left_mid, mid_mid, right_mid_new = make_nonuniform_row(outer[1], 0.7)

# Left pair
ax_left_mid1 = fig.add_subplot(left_mid[0, 0])
ax_left_mid1.set_box_aspect(2)
ax_left_mid1.set_xticks([]); ax_left_mid1.set_yticks([])
ax_left_mid1.set_title('Sound Index')
axes.append(ax_left_mid1)

ax_left_mid2 = fig.add_subplot(left_mid[0, 1])
ax_left_mid2.set_box_aspect(2)
ax_left_mid2.set_xticks([]); ax_left_mid2.set_yticks([])
#ax_left_mid2.set_title('Sound Index')
axes.append(ax_left_mid2)

# Middle existing subplot
ax_right_mid = fig.add_subplot(mid_mid[0, 0])
ax_right_mid.set_box_aspect(1)
ax_right_mid.spines['top'].set_visible(False)
ax_right_mid.spines['right'].set_visible(False)
ax_right_mid.set_ylabel('|Modulation Index|',labelpad=0.7)
ax_right_mid.tick_params(direction='in', labelsize=size, length=1)
axes.append(ax_right_mid)

# NEW rightmost subplot (middle row)
ax_new_mid = fig.add_subplot(right_mid_new[0, 0])
ax_new_mid.set_box_aspect(1)
ax_new_mid.spines['top'].set_visible(False)
ax_new_mid.spines['right'].set_visible(False)
#ax_new_mid.set_title('New Middle', fontsize=7)
ax_new_mid.set_xlabel('Context input (V/s)')
ax_new_mid.set_ylabel('Sound modulation index', labelpad=2)
ax_new_mid.tick_params(direction='in', labelsize=size, length=1)
axes.append(ax_new_mid)

# ── 5️⃣ Bottom row: also 4 subplots ───────────────────────────
left_bot, mid_bot, right_bot_new = make_nonuniform_row(outer[2], 0.7)

# Left pair
ax_left_bot1 = fig.add_subplot(left_bot[0, 0])
ax_left_bot1.set_box_aspect(2)
#ax_left_bot1.set_title('Photostim'+'\n'+ 'Index')
ax_left_bot1.set_xticks([]); ax_left_bot1.set_yticks([])
axes.append(ax_left_bot1)

ax_left_bot2 = fig.add_subplot(left_bot[0, 1])
ax_left_bot2.set_box_aspect(2)
ax_left_bot2.set_title('Photostim'+'\n'+ 'Index')
ax_left_bot2.set_xticks([]); ax_left_bot2.set_yticks([])
axes.append(ax_left_bot2)

# Middle existing subplot
ax_right_bot = fig.add_subplot(mid_bot[0, 0])
ax_right_bot.set_box_aspect(1)
ax_right_bot.set_ylabel('|Modulation Index|', labelpad=0.7)
ax_right_bot.spines['top'].set_visible(False)
ax_right_bot.spines['right'].set_visible(False)
ax_right_bot.tick_params(direction='in', labelsize=size, length=1)
axes.append(ax_right_bot)

# NEW rightmost subplot (bottom row)
ax_new_bot = fig.add_subplot(right_bot_new[0, 0])
ax_new_bot.set_box_aspect(1)
ax_new_bot.spines['top'].set_visible(False)
ax_new_bot.spines['right'].set_visible(False)
#ax_new_bot.set_title('New Bottom', fontsize=7)
ax_new_bot.set_xlabel('Photostim input (V/s)')
ax_new_bot.set_ylabel('Photostim modulation index', labelpad=2)
ax_new_bot.tick_params(direction='in', labelsize=size, length=1)
axes.append(ax_new_bot)

plt.show()
```

```python
#### """ rate plots """
from matplotlib.ticker import FormatStrFormatter
from scipy.ndimage import gaussian_filter
xticks = [-0.1, 0.1, 0.3]
yticks_upper = [1,2,3,4]
yticks_lower = [1,3,5]
for i in range(2):
    for j in range(3):
        ax = fig.add_subplot(nested_top[i, j])
        ax.plot(np.linspace(-100, 500, 600)/1000, gaussian_filter(pop_sound_rates_act[3*i+j][900:1500], sigma=4)*1000, color='rebeccapurple', label='Sound, active', linewidth=1)
        ax.plot(np.linspace(-100, 500, 600)/1000, gaussian_filter(pop_sound_rates_pass[3*i+j][900:1500], sigma=4)*1000, color='rebeccapurple', label='Sound, passive', alpha=0.4, linewidth=1)
        ax.plot(np.linspace(-100, 500, 600)/1000, gaussian_filter(pop_stim_rates_act[3*i+j][900:1500], sigma=4)*1000, color='orange', label='Photostim + sound, active',linewidth=1)
        ax.plot(np.linspace(-100, 500, 600)/1000, gaussian_filter(pop_stim_rates_pass[3*i+j][900:1500], sigma=4)*1000, color='orange', label='Photostim + sound, passive', alpha=0.4, linewidth=1)
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.tick_params(direction='in', length=1)
        ax.set_xticks(xticks)
        ax.xaxis.set_major_formatter(FormatStrFormatter('%.1f'))
        if i == 0:
            ax.vlines(0, ymin=-0.001, ymax=4, linestyles='solid', color='gray')
            ax.set_ylim(0.5, 4)
            ax.set_xlim(-0.1, 0.3)
            ax.set_yticks(yticks_upper)
            ax.yaxis.set_major_formatter(FormatStrFormatter('%.0f'))
        if i == 1: 
            ax.vlines(0, ymin=-0.001, ymax=6, linestyles = 'solid', color='gray')
            ax.set_ylim(0.5, 5)
            ax.set_xlim(-0.1, 0.3)
            ax.set_yticks(yticks_lower)
            ax.yaxis.set_major_formatter(FormatStrFormatter('%.0f'))
        ax.set_box_aspect(1)    
        axes.append(ax)
        


"""sound MI - defining all quantities"""
normalised = 'no'
mi_sound_pass_norm = np.loadtxt(file_path/"normalized_mi_sound_pass.txt")
mi_sound_act_norm = np.loadtxt(file_path/"normalized_mi_sound_act.txt")
mi_sound_pass_no_denom = np.loadtxt(file_path/"mi_sound_pass_no_denom.txt")
mi_sound_act_no_denom = np.loadtxt(file_path/"mi_sound_act_no_denom.txt")

ne1, ne2, ne_non, ni1, ni2, ni_non = 2040, 4080, 6800, 7160, 7520, 8000
if normalised == 'yes':
    mi_sound_pass = mi_sound_pass_norm
    mi_sound_act = mi_sound_act_norm

if normalised == 'no':
    mi_sound_pass = mi_sound_pass_no_denom
    mi_sound_act = mi_sound_act_no_denom

act_pass_avg = (mi_sound_act + mi_sound_pass)/2


e1, e2, enon, i1, i2, inon = act_pass_avg[:ne1], act_pass_avg[ne1:ne2], act_pass_avg[ne2:ne_non], act_pass_avg[ne_non:ni1], act_pass_avg[ni1:ni2], act_pass_avg[ni2:]

sort_avg_arg = np.concatenate( (np.argsort(e1),
(ne1) + np.argsort(e2),
(ne2) + np.argsort(enon),
(ne_non) + np.argsort(i1),
(ni1) + np.argsort(i2),
(ni2) + np.argsort(inon) ) )

mi_pass = mi_sound_pass[sort_avg_arg]
mi_act = mi_sound_act[sort_avg_arg]

# ── 3️⃣ middle row: 3 subplots (aspect 1.5) ───────────────────
""" sound MI for excitatory """
import matplotlib.colors as mcolors
vals_E = np.concatenate([mi_act[:ne_non], mi_pass[:ne_non]])
norm_E = mcolors.Normalize(vmin=-4, vmax=4)
cmap = plt.get_cmap('bwr')  # Diverging colormap (red to blue)

# Plot horizontal lines for E
for i in range(ne_non):
    ax_left_mid1.hlines(y=i, xmin=0, xmax=1, color=cmap(norm_E(mi_act[i])), linewidth=1)
    ax_left_mid1.hlines(y=i, xmin=1, xmax=2, color=cmap(norm_E(mi_pass[i])), linewidth=1)

ax_left_mid1.hlines(y=ne1, xmin=0, xmax=2, color='k', linewidth=1)

ax_left_mid1.hlines(y=ne2, xmin=0, xmax=2, color='k', linewidth=1)


sm_E = plt.cm.ScalarMappable(cmap=cmap, norm=norm_E)
#fig.colorbar(sm_E, ax=ax_left_mid1, orientation="vertical", shrink=1).ax.tick_params(labelsize=7, direction='in', length=1)
 # 
ax_left_mid1.margins(0)

""" Sound MI for inhibitory """
from mpl_toolkits.axes_grid1 import make_axes_locatable

vals_I = np.concatenate([mi_act[ne_non:], mi_pass[ne_non:]])
norm_I = mcolors.Normalize(vmin=-4, vmax=4)
for i in range(ne_non, 8000):
    ax_left_mid2.hlines(y=i, xmin=0, xmax=1, color=cmap(norm_I(mi_act[i])), linewidth=1)
    ax_left_mid2.hlines(y=i, xmin=1, xmax=2, color=cmap(norm_I(mi_pass[i])), linewidth=1)

sm_I = plt.cm.ScalarMappable(cmap=cmap, norm=norm_I)

ax_left_mid2.hlines(y=ni1, xmin=0, xmax=2, color='k', linewidth=1)
ax_left_mid2.hlines(y=ni2, xmin=0, xmax=2, color='k', linewidth=1)

ax_left_mid2.margins(0)

# 1) Freeze current layout so bbox is final
fig.canvas.draw()

# 2) Compute a cbar axes that exactly matches the subplot height
bbox = ax_left_mid2.get_position()  # in figure coordinates (0..1)
pad = 0.008        # gap to the right of the subplot (in fig coords)
cbar_w = 0.015     # colorbar width (in fig coords)

# If space is tight, you can reduce pad or cbar_w
cax = fig.add_axes([bbox.x1 + pad, bbox.y0, cbar_w, bbox.height])

# 3) Draw colorbar in that fixed axes
cb = fig.colorbar(sm_I, cax=cax, orientation='vertical')
cb.ax.tick_params(direction='in', labelsize=size, length=1)
cb.ax.yaxis.set_major_formatter(FormatStrFormatter('%.0f'))
cb.set_label('Modulation Index', fontsize=size, labelpad=8, rotation=270)



#SCATTER PLLOTS FOR ALL E AND ALL I
mean_E_pass = np.mean(np.abs(mi_pass[:ne_non]))
std_E_pass = np.std(np.abs(mi_pass[:ne_non]))
mean_I_pass = np.mean(np.abs(mi_pass[ne_non:]))
std_I_pass = np.std(np.abs(mi_pass[ne_non:]))

mean_E_act = np.mean(np.abs(mi_act[:ne_non]))
std_E_act = np.std(np.abs(mi_act[:ne_non]))
mean_I_act = np.mean(np.abs(mi_act[ne_non:]))
std_I_act = np.std(np.abs(mi_act[ne_non:]))

errorE = np.array([std_E_act, std_E_pass])/np.sqrt(ne_non)   # sem
errorI = np.array([std_I_act, std_I_pass])/np.sqrt(ni_non-ne_non)  # sem
# Set x-ticks and labels
ticks = np.arange(4, 20, 4)

ax_right_mid.tick_params(direction='in', labelsize=size, length=1)
ax_right_mid.set_xticks(ticks)
ax_right_mid.set_ylim(0, 1.5)
yticks = [0., 0.5, 1.0, 1.5]
labels = ['Active', 'Passive', 'Active', 'Passive']
ax_right_mid.set_xticklabels(labels, fontname='Arial', fontsize=size, rotation=45)
ax_right_mid.yaxis.set_major_formatter(FormatStrFormatter('%.1f'))
ax_right_mid.set_yticks(yticks)

# Scatter points
ax_right_mid.scatter([ticks[0], ticks[1]], [mean_E_act, mean_E_pass], s=7, color='green')
ax_right_mid.scatter([ticks[2], ticks[3]], [mean_I_act, mean_I_pass], s=7, color='red')
ax_right_mid.errorbar([ticks[0], ticks[1]], [mean_E_act, mean_E_pass], yerr = errorE, color='green', capsize=5)
ax_right_mid.errorbar([ticks[2], ticks[3]], [mean_I_act, mean_I_pass], yerr=errorI, color='red', capsize=5)

# Connecting lines
ax_right_mid.plot([ticks[0], ticks[1]], [mean_E_act, mean_E_pass], color='green', label='Excitatory')
ax_right_mid.plot([ticks[2], ticks[3]], [mean_I_act, mean_I_pass], color='red', label='Inhibitory')

ax_new_mid.scatter(I_cont, mi_sound_pass_no_denom[contextID-1], s=1, color='thistle')
ax_new_mid.scatter(I_cont, mi_sound_act_no_denom[contextID-1], s=1, color='rebeccapurple')
ax_new_mid.set_xlim(0, 20)
ax_new_mid.set_ylim(-4, 8)
ax_new_mid.set_yticks([-4, 0, 4, 8])
```
